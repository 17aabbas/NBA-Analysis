```python
import pandas as pd
filepath = '/content/sample_data/nba.csv'

df = pd.read_csv(filepath)
df.head()
```

```python
df = df.drop(['RANK', 'TEAM_ID', 'TEAM', 'PLAYER'], axis = 1)
df['Season_type'] = df['Season_type'].apply(lambda x: 0 if x=='Regular%20Season' else 1)
df[['Start Year', 'End Year']] = df['year'].str.split('-', expand = True)
df['Start Year'] = pd.to_datetime(df['Start Year'])
df['End Year'] = pd.to_datetime('20' + df['End Year'])
df['Start Year'] = df['Start Year'].dt.year
df['End Year'] = df['End Year'].dt.year
df = df.drop(['year', 'End Year'], axis = 1)
df['PPG'] = df['PTS']/df['GP']
float_columns = ['FG_PCT','FG3_PCT','FT_PCT', 'PPG']
df[float_columns] = df[float_columns].round(2)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['Start Year','Season_type', 'PLAYER_ID', 'GP', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M',
       'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST',
       'STL', 'BLK', 'TOV', 'PF', 'PTS', 'EFF', 'AST_TOV', 'STL_TOV', 'PPG']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (25,15))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop('PPG', axis = 1)
y = df['PPG']

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size = 0.2, random_state = 42)
model = LinearRegression()
model.fit(X_train, y_train)

print(f"Model R^2 score: {model.score(X_test, y_test)}")
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print(f"Cross-validation scores: {scores}")
print(f"Mean R^2 score: {scores.mean()}")
```
